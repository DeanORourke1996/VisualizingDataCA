<h1>1. Introduction</h1>
<h3>Overview</h3>
<p>These datasets are provided by the CDC and are a part of continous state-based (US) surveillance systems that collect information about modifiable risk factors for chronic diseases. There are multiple indicators to provide a broad picture of the public health burden of Cardiovascular Diseases and other associated risk factors.</p>

```r
# Read into dataframe 1
df1 <- read.csv('./data/BHRF_CDC_InterviewSurvey_CDC.csv', stringsAsFactors=FALSE)
# Read into dataframe 2
df2 <- read.csv('./data/BHRF_CDC_Cardiovascular_Disease.csv', stringsAsFactors=FALSE)
```

<h3>Peek at data</h3>
<p>Quick glance at the datasets</p>

```r
head(df1, n=2)
head(df2, n=2)
```

<h1>2. Clean Datasets and Merge</h1>
<p>Sets equal column names for the datasets before merging</p>

```r
names(df2) = names(df1)
```

<p>Merge & transpose datasets using rbind()</p>

```r
DF <- rbind(df1, df2)
```

<h3>2.1 Analyze Missing Data</h3>

```r
missingVals <- colSums(is.na(DF))
```

```r
par(mar=c(5,15,1,0))
barplot(missingVals[missingVals != 0],
    main = "Missing Values",
    xlab = "",
    names.arg = names(missingVals[missingVals != 0]),
    col = "blue",
    las = 1,
    horiz = TRUE)
```

<h3>2.2 Remove Missing Values</h3>

```r
DF <- DF %>%
    na.omit()
dim(DF)
```

<h1>3. Data Extraction and Seperation</h1>
<p>Collect numerical columns</p>

```r
numeric <- unlist(lapply(DF, is.numeric))
numerical <- DF[, numeric]
```

<h3>3.1 Check for Primary Key within Dataset</h3>
<p>LocationID is not a unique column</p>

```r
distinct(DF, Year, LocationID)
select(DF[which(DF$Year==2000),], Year, LocationID)

# num of rows
# nrow(distinct(DF, Year, LocationID))
# nrow(select(DF[which(DF$Year==2000),], Year, LocationID))
```

<h3>3.2 Add Foreign Key to Dataset</h3>
<p>No unique key for the dataset so will add one</p>

```r
DF <- tibble::rowid_to_column(DF, "pk")
```

<h3>3.3 Outliers and Redundant Data</h3>
<p>Here we can see that 2015, 2013 and 2011 provide the dataset with the most amount of records and highest confidence upper bound (>95%). It is appropriate to drop the other years as the Lower Confidence Upper Bounds is higher in those years ('18, '17, '16, '14 & '12 as is the Higher Confidence Upper Bounds</p>

```r
DF %>%
    ggplot() +
    geom_point(aes(DF$HighConfidenceLimit, DF$Year, colour='red')) +
    geom_point(aes(DF$LowConfidenceLimit, DF$Year, colour='blue'))
```

```r
hist(DF$Year, col="red", xlab="Year", main="Histogram of Year")
```

<p>Another representation of the frequency of values per year</p>

```r
hist(DF$Data_Value, col="orange", xlab="Data Value", main="Histogram of Data Value")
```

<p>Positively skewed data value to frequency chart. Once data from specified years is isolated, this should have much more normal distribution</p>

```r
DF <- subset(DF, Year==2015 | Year==2013 | Year==2011)
```

<p>Now to have another look</p>

```r
hist(DF$Data_Value, col="orange", xlab="Data Value", main="Histogram of Data Value")
```

<p>The skew is not as severe and the distribution is slightly more normalised with much lesser disparity in Frequency range</p>

<p>For the purposes of this research, the LocationID and GeoLocation fields are not necessary as they will not be used. Hence they can be removed</p>

```r
DF <- subset(DF, select = -c(LocationID, GeoLocation))
```

<p>Gather up-to-date numerical columns again</p>

```r
numeric <- unlist(lapply(DF, is.numeric))
numerical <- DF[, numeric]
```

<h1>4. Data Visualisation</h1>

<h3>Correlation Graph on Numerical Columns</h3>
<p>First we will look at a correlation graph between numerical columns</p>

```r
numerical.cor = cor(numerical)
corrplot(numerical.cor)
```

<h3>Cardiovascular Diseases Amongst Various Age Groups</h3>

<p>First seperate ages into subsets using the subset function</p>

```r
ages <- subset(DF, Break_Out_Category == "Age")
```

```r
ggplot(ages, aes(x = Break_Out)) +
    geom_bar(width=0.6, fill="orange") +
    geom_text(stat='count', aes(label=stat(count)), vjust=-0.6) +
    labs(title = "Cardiovascular Diseases Amongst Various Age Groups", x = "Age Groups", y = "Frequency") +
    theme_classic()
```

<h3>Most Significant Causes of Cardiovascular Diseases</h3>

```r
cvds <- subset(DF, Topic='*')  # Get all CVDs
unique_cvds <- unique(cvds$Topic)  # Unique Values (11)
cvd_stats <- count(cvds, cvds$Topic)  # Get Totals of Each CVD
cvd_vectored <- as.vector(cvd_stats[,2])  # Vectorize to be used in plot
pie(cvd_vectored, unique_cvds)
```

<h3>Cardiovascular Diseases Among Genders</h3>

<p>Categorise Incidents by their Breakout Category, indicating Gender</p>

```r
gender <- subset(DF, Break_Out_Category == "Gender")
```

```r
ggplot(gender, aes(x = Break_Out, fill = Break_Out)) +
    geom_bar(position = position_dodge()) +
    labs(title = "Cardiovascular Diseases Amongst Genders", x = "Gender", y = "Frequency") +
    geom_text(stat='count',
             aes(label=stat(count)),
             position=position_dodge(width=1),
             vjust=-0.5)+
        theme_classic()
```

<p>Essentially identitcal incidence rate among males and females</p>

<h3>State of Pennsylvania by Disease Incidence</h3>

```r
p <- DF %>% 
    filter(LocationDesc=="Pennsylvania") %>%
    ggplot(aes(x=Topic))+
        geom_bar(fill="blue", position=position_dodge()) +
        labs(title="Cardiovascular Diseases in Pennsylvania", x="Cardiovascular Disease", y="Frequency") +
        geom_text(stat='count',
                 aes(label=stat(count)),
                 position=position_dodge(width=1),
                 vjust=-0.5)+
        theme(text = element_text(size = 25), element_line(size = 0.1))

# This plot is quite large so here the repr package will be used to modify the plots display
# Another package "scales" is used to scale and wrap the text in the x and y axes.
options(repr.plot.width = 20, repr.plot.height = 10)
p + scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
```

<p>Obesity has is the leading cause of Cardiovascular Disease in Pennsylvania</p>

<h3>Comparison to State of Florida</h3>

```r
DF %>% 
    filter(LocationDesc=="Florida") %>%
        ggplot(aes(x=Topic))+
        geom_bar(fill="red", position=position_dodge()) +
        labs(title="Cardiovascular Diseases in Florida", x="Cardiovascular Diseases", y="Frequency") +
        geom_text(stat='count',
                 aes(label=stat(count)),
                 position=position_dodge(width=1),
                 vjust=-0.5)+
        theme(text = element_text(size = 20), element_line(size = 0.1)) + 
        scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
```

<p>Similarly, obesity is the number one cause of Cardiovascular disease in Florida</p>

<h3>Cause of Disease and Incidence Among 18 to 24 years olds</h3>

```r
DF %>% 
    filter(Break_Out=="18-24") %>%
        ggplot(aes(x=Topic))+
        geom_bar(fill="orange", position=position_dodge()) +
        labs(title="Incidence Rates and Causes Among 18-24 year Olds", x="Cardiovascular Diseases", y="Frequency") +
        geom_text(stat='count',
                 aes(label=stat(count)),
                 position=position_dodge(width=1),
                 vjust=-0.5)+
        theme(text = element_text(size = 30), element_line(size = 0.2)) + scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
```

<p>This graph shows that young adults get Cardiovascular Disease for much different reasons with the highest being nutrition related</p>

<h3>Compared to Cause and Incidence Among 25-44 year olds</h3>

<p>Interestingly, the cause of incidence in the immediate range above 18-24 year olds changes significantly as reasons fall back inline with the overall story the data tells: Obseity being the main cause of Cardiovascular Disease</p>

```r
DF %>%
    filter(Break_Out=="25-44") %>%
        ggplot(aes(x=Topic)) +
        geom_bar(fill="lightblue", position=position_dodge()) +
        labs(title="Incidence Rates and Causes Among 25-44 year Olds", x="Cardiovascular Diseases", y="Frequency") +
        geom_text(stat='count',
                 aes(label=stat(count)),
                 position=position_dodge(width=1),
                 vjust=-0.5)+
        theme(text = element_text(size = 30), element_line(size = 0.2)) + scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
```

<h3>65 and Over</h3>

<p>The trend continues as the age increases to 65s and over. This is where a significant amount of the data derives from. Cholesterol and Hypertension also follow closely to form the top 3 overall causes of Cardiovascular Disease</p>

```r
DF %>%
    filter(Break_Out=="65+") %>%
        ggplot(aes(x=Topic)) +
        geom_bar(fill="lightgreen", position=position_dodge()) +
        labs(title="Incidence Rates and Causes Over 65s", x="Cardiovascular Diseases", y="Frequency") +
        geom_text(stat='count',
                 aes(label=stat(count)),
                 position=position_dodge(width=1),
                 vjust=-0.5)+
        theme(text = element_text(size = 20), element_line(size = 0.2)) + scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
```

<h3>75 and Over</h3>

```r
DF %>%
    filter(Break_Out=="75+") %>%
        ggplot(aes(x=Topic)) +
        geom_bar(fill="lightpink", position=position_dodge()) +
        labs(title="Incidence Rates and Causes Over 75s", x="Cardiovascular Diseases", y="Frequency") +
        geom_text(stat='count',
                 aes(label=stat(count)),
                 position=position_dodge(width=1),
                 vjust=-0.5)+
        theme(text = element_text(size = 20), element_line(size = 0.2)) + scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
```

<h3>A Look at 3 Age Groups in a Stacked Bar (18-24, 35+ and 65+)</h3>

<p>We can see from the 35+ age group that there is a sharp increase in the Obesity factor</p>
<p>Inversely there is a sudden decrease in Hypertension between the youngest demographic and the middle range</p>

```r
over18a65 <- DF %>% filter(Break_Out=="18-24" | Break_Out=="35+" | Break_Out=="65+")
```

```r
over18a65_plot <- over18a65 %>%
        ggplot(aes(x=Topic, fill=Break_Out)) +
        geom_bar(position="stack", stat="count") +
        labs(title="A Comparison between three age groups", x="Cardiovascular Diseases", y="Frequency") +
        theme(text = element_text(size = 20), element_line(size = 0.2)) + scale_x_discrete(labels = function(x) lapply(strwrap(x, width=5, simplify=FALSE), paste, collapse='\n'))
                                                                                           
over18a65_plot
```
